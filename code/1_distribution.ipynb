```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.concat([pd.read_csv('../raw_data/../raw_data/yellow_tripdata_2018-06.csv'),
                pd.read_csv('../raw_data/../raw_data/yellow_tripdata_2018-07.csv')])
df.dropna(inplace=True)
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'], errors='coerce')
df['trip_time_cost'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.seconds.astype(float) / 60
```

```python
sns.distplot(df['trip_distance'])
plt.title("Distribution of trip distance")
plt.xlabel('trip distance (miles)')
plt.ylabel("count")
plt.savefig('../plots/trip_distance.png')
plt.show()
```

```python
sns.distplot(df['tip_amount'])
plt.title("Distribution of tip amount")
plt.xlabel('tip amount (USD)')
plt.ylabel("count")
plt.savefig('../plots/tip_amount.png')
plt.show()
```

```python
sns.distplot(df['fare_amount'])
plt.title("Distribution of fare amount")
plt.xlabel('fare amount (USD)')
plt.ylabel("count")
plt.savefig('../plots/fare_amount.png')
plt.show()
```

```python
sns.distplot(df['trip_time_cost'])
plt.title("Distribution of trip time cost")
plt.xlabel('trip time cost (min)')
plt.ylabel("count")
plt.savefig('../plots/trip_time_cost.png')
plt.show()
```
